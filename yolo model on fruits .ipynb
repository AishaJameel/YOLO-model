```python
from ultralytics import YOLO
import cv2

# Load your trained YOLO model
model = YOLO(r"C:\Users\amnaj\deep learning\runs\detect\train9\weights\best.pt")

# Load a video file
video_path = r"C:\Users\amnaj\Downloads\4403833-hd_1920_1080_30fps.mp4"  # <-- change this to your video path
cap = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties for saving output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output.mp4", fourcc, 20.0,
                      (int(cap.get(3)), int(cap.get(4))))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame, stream=True)

    # Draw detections on the frame
    for r in results:
        annotated_frame = r.plot()

    # Show and save the frame
    cv2.imshow("YOLO Detection", annotated_frame)
    out.write(annotated_frame)
    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
results = model(frame, imgsz=640, conf=0.1, stream=True)


```

```python
import cv2
from ultralytics import YOLO
import easyocr

# Load your trained YOLO model (path to best.pt)
model = YOLO(r"C:\Users\amnaj\deep learning\runs\detect\train9\weights\best.pt")

# Initialize OCR reader (English)
reader = easyocr.Reader(['en'])

# Load your video file (or webcam: use 0)
video_path = r"C:\Users\amnaj\Downloads\13035271_1920_1080_24fps.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("❌ Error: Could not open video.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection on the frame
    results = model(frame)

    # Loop through all detected boxes
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])

            # Draw detection box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Crop the detected plate region
            plate_crop = frame[y1:y2, x1:x2]

            # OCR on the cropped image
            ocr_result = reader.readtext(plate_crop, detail=0)
            if len(ocr_result) > 0:
                text = " ".join(ocr_result)
                cv2.putText(frame, text, (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)
                print(f"📖 Detected Plate Text: {text}")

    cv2.imshow("YOLO + OCR License Plate Detection", frame)

    # Exit with 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
